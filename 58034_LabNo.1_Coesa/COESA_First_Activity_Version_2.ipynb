```python
grades_df = pd.read_excel('/content/class_grades.xlsx') 
```

```python
# Data Preprocessing
## Replace NaN Values with 0

grades_df = grades_df.fillna(0)
grades_df
```

```python
# Gets the value of sem grade for the dataframe
def get_grades(df): 
  df['semestral_grades'] = (df['prelim_grades']*.30) + (df['midterm_grades']*.30) + (df['final_grades']*.40) # Computes the semestral grades dataframe
  return df # Return to the get grades dataframe

get_grades(grades_df)
```

```python
def get_class_stats(df):
  # Get all the grades column from the dataframe
  df = df.fillna(0)
  df['semestral_grades'] = (df['prelim_grades']*.30) + (df['midterm_grades']*.30) + (df['final_grades']*.40) # Computes the semestral grades dataframe
  grades = df[['prelim_grades', 'midterm_grades', 'final_grades', 'semestral_grades']]
  

  def max_min(grades):
    return grades.apply(lambda s: pd.Series([s.min(), s.max()],index=['Lowest', 'Highest'])) # Create a New Dataframe containing Highest and Lowest Grades


  def clt(df):
    grades_transp = df.filter(['prelim_grades', 'midterm_grades', 'final_grades', 'semestral_grades'], axis=1)
    grades_transp = grades_transp.T
    # Declare Dictionary to store the CLT Values then Declare it as a DataFrame
    stats = {'Mean': [], 'Median': [], 'Mode': [], 'St dev': []}
    clt_df = pd.DataFrame(stats)

    clt_df['Mean'] = grades_transp.mean(axis=1)
    clt_df['Median'] = grades_transp.median(axis=1)
    clt_df['Mode'] = grades_transp.mode(axis=1)
    clt_df['St dev'] = grades_transp.std(axis=1)
    return clt_df
  
  print("Lowest and Highest Grades Dataframe: \n", max_min(grades),'\n\n')
  print('CLT DataFrame: \n', clt(df),'\n\n')
```

```python
get_class_stats(grades_df)
grades_df
```

####Get similar students function
Creating a function that takes an input of two (2) dataframes and then compare each other. The function will return a list of common students wrt ID number that is detected in both dataframe.

```python
def get_similar(df_class1, df_class2):
  similar_df = pd.merge(left=df_class1, right=df_class2, left_on='Student_no.:', right_on='Student_no.:')
  stud_list = similar_df['Student_no.:'].tolist()
  return stud_list
```

####Creating class section
Creating a class "section" will include functions from tasks 2 and 3 as its methods, such as function max_min and clt. Additionally, two methods are added, namely get_failed and fail_count, that will filter the student who got a semester grade lower than 70 and count their number. This method will return all necessary variables that are needed for each function.

```python
grades_df = pd.read_excel('/content/class_grades.xlsx') 
```

```python
class section:
  def __init__(self, df):
    self.df = df
    self.preprocess()
    self.df['semestral_grades'] =  (self.df['prelim_grades']*.30) + (self.df['midterm_grades']*.30) + (self.df['final_grades']*.40)
    self.grades_col = self.df[['prelim_grades', 'midterm_grades', 'final_grades', 'semestral_grades']]

    
  def preprocess(self):
    self.df = self.df.fillna(0)
    self.df['name'] = '*****'
  
  
  def get_df(self):
    return self.df 


  def max_min(self):
    return self.grades_col.apply(lambda grade: pd.Series([grade.min(), grade.max()],index=['Lowest', 'Highest']))
  

  def clt(self):
    self.grades_transp = self.df.filter(['prelim_grades', 'midterm_grades', 'final_grades', 'semestral_grades'], axis=1)
    self.grades_transp = self.grades_transp.T
    # Declare Dictionary to store the CLT Values then Declare it as a DataFrame
    self.stats = {'Mean': [], 'Median': [], 'Mode': [], 'St dev': []}
    self.clt_df = pd.DataFrame(self.stats)
    # Acquire CLT using functions
    self.clt_df['Mean'] = self.grades_transp.mean(axis=1)
    self.clt_df['Median'] = self.grades_transp.median(axis=1)
    self.clt_df['Mode'] = self.grades_transp.mode(axis=1)
    self.clt_df['St dev'] = self.grades_transp.std(axis=1)
    return self.clt_df
    

  def get_failed(self):
    self.df['Remark'] = np.where(self.df['semestral_grades'] < 70, 'Failed', 'Passed')
    self.failed_df = self.df[(self.df['Remark']=='Failed')]
    return self.failed_df


  def fail_count(self):
    return "Total no. of Failed Students: " + str(len(self.failed_df))
```
