```python
# Data Preprocessing
## Replace NaN Values with 0

grades_df = grades_df.fillna(0)
grades_df
```

```python
# Gets the value of the dataframe
def get_grades(df): 
  df['semestral_grades'] = (df['prelim_grades']*.30) + (df['midterm_grades']*.30) + (df['final_grades']*.40) # Computes the semestral grades dataframe
  return df # Return to the get grades dataframe

get_grades(grades_df)
```

```python
def get_class_stats(df):
  # Get all the grades column from the dataframe
  grades = df[['prelim_grades', 'midterm_grades', 'final_grades', 'semestral_grades']]

  def max_min(grades):
    return grades.apply(lambda s: pd.Series([s.min(), s.max()],index=['Lowest', 'Highest'])) # Create a New Dataframe containing Highest and Lowest Grades

  def clt(grades):
    df['Mean'] = grades.mean(axis=1)
    df['Median'] = grades.median(axis=1)
    # A for loop is used to display the mode or its absence
    unique_grades = []
    for grade in grades:
      if grade not in unique_grades:
        unique_grades.append(grade)
    if len(set(unique_grades)) == 4:
      df['Mode'] = 'NaN'
    else: 
      df['Mode'] = grades.mode(axis=1)
    df['St dev'] = grades.std(axis=1)

  print("Lowest and Highest Grades Dataframe")
  print(max_min(grades))
  clt(grades)
 
get_class_stats(grades_df)
grades_df
```

####Creating class section
Creating a class "section" will include functions from tasks 2 and 3 as its methods, such as function max_min and clt. Additionally, two methods are added, namely get_failed and fail_count, that will filter the student who got a semester grade lower than 70 and count their number. This method will return all necessary variables that are needed for each function.

```python
class section:
  def __init__(self, df):
    self.df = df
    self.df['semestral_grades'] =  (self.df['prelim_grades']*.30) + (self.df['midterm_grades']*.30) + (self.df['final_grades']*.40)
    self.grades_col = df[['prelim_grades', 'midterm_grades', 'final_grades', 'semestral_grades']]

  def max_min(self):
    return self.grades_col.apply(lambda grade: pd.Series([grade.min(), grade.max()],index=['Lowest', 'Highest']))
  
  def clt(self):
    self.df['Mean'] = self.grades_col.mean(axis=1)
    self.df['Median'] = self.grades_col.median(axis=1)
    # A for loop is used to display the mode or its absence
    unique_grades = []
    for grade in self.grades_col:
      if grade not in unique_grades:
        unique_grades.append(grade)
    if len(set(unique_grades)) == 4:
      self.df['Mode'] = 'NaN'
    else: 
      self.df['Mode'] = self.grades_col.mode(axis=1)
    self.df['St dev'] = self.grades_col.std(axis=1)
    return self.df
    
  def get_failed(self):
    self.df['Remark'] = np.where(self.df['semestral_grades'] < 70, 'Failed', 'Passed')
    self.failed_df = self.df[(self.df['Remark']=='Failed')]
    return self.failed_df

  def fail_count(self):
    counter = len(self.failed_df)
    return "Total no. of Failed Students: " + str(counter) 

```

```python
class_data = section(grades_df)
print(class_data.max_min())
```
